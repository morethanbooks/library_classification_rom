```python
for year_path in years_paths[-42:-2]:

    year_name  = os.path.splitext(os.path.split(year_path)[1])[0]
    print(year_name)

    df = pd.read_parquet(year_path)

    df = extract_data_from_df.extract_romance_studies(df)

    df = df.loc[df["romance_studies"] == 1]

    df = df.groupby("ppn").head(1)

    df.index = df.ppn

    df.drop(columns = ["ppn"], inplace=True)


    df.to_parquet(outdir + year_name + ".parquet")
    

```
